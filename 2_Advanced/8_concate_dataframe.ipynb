```python
import pandas as pd 
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading the data
dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

# converting to the right time value

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
```

```python
df.sample(5)
```

```python
months = df['job_posted_month'].unique()
```

```python
dict_month = {month: df[df['job_posted_month'] == month] for month in months}

dict_month['Jan']
```

```python
Q_1 = pd.concat([dict_month['Jan'], dict_month['Feb'], dict_month['Mar']], ignore_index= True)
```

```python
Q_1['job_posted_month'].value_counts().plot(kind= 'bar', figsize= (10,6), color = 'pink')
```

```python
import seaborn as sns

sns.barplot( x= Q_1['job_posted_month'].value_counts().index, y= Q_1['job_posted_month'].value_counts().values, palette= 'viridis')
```

```python
Q_1['job_posted_month'].value_counts()
```

```python
#saving the file
Q_1.to_csv('quarter1.csv') 
```
